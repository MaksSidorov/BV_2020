```python
import pandas as pd
import numpy as np
```

# Загрузка данных

```python
df_z = pd.read_excel('data/DATA.xlsx', sheet_name='Завалки')
```

```python
df_p = pd.read_excel('data/DATA.xlsx', sheet_name='Проверка')
```

```python
df_r =  pd.read_excel('data/DATA.xlsx', sheet_name='Рулоны')
```

```python
df_v =  pd.read_excel('data/DATA.xlsx', sheet_name='Валки')
```

# Преобразование данных

## Материал валка


```python
df_v.head()
```

```python
arr = df_v['материал_валка'].values
```

```python
for i in range(len(arr)):
    arr[i] = int(arr[i][14:])
```

```python
df_v['материал_валка'] = arr
```

## Завалки и Проверка(Основная часть обучающей выборки)


```python
df_z.head()
```

```python
df_t = df_z.append(df_p)
```

```python
df_t.head(20)
```

```python
df_t['положение_в_клети'] = df_t['положение_в_клети'] == 'верх'
df_t['положение_в_клети'] = df_t['положение_в_клети'].astype('int')
```

```python
df_t['время'] = (df_t['дата_вывалки'] - df_t['дата_завалки']).astype('int64')/ 1000000000
```

```python
rul_ids = dict()
v = []
c = -1
for i in df_t['дата_завалки']:
    if str(i) not in rul_ids.keys():
        rul_ids[str(i)] = c
        c += 1
    v.append(c)
df_t['rul_id'] = v
```

## Рулоны


```python
df_r.head()
```

```python
times = np.array([])
for i in df_t['дата_вывалки']:
    if i not in times:
        times = np.append(times, i)
```

```python
len(times)
```

```python
ids = []
c = 0
for i in df_r['Время_обработки']:
    if c < len(times) and i > times[c]:
        c+=1
    ids.append(c)
df_r['rul_id'] = ids
```

```python
times = []
for i in df_t['дата_вывалки'].values.tolist():
    if i not in times:
        times.append(int(i))
```

```python
df_r = df_r.loc[df_r['rul_id'] < len(times)]
```

```python
time = []
for i in range(len(times)):
    lst = df_r.loc[df_r['rul_id'] == i]['Время_обработки'].values.tolist()
    lst.append(times[i])
    for j in range(1, len(lst)):
        time.append((lst[j] - lst[j-1]) / 1000000000)
df_r['time'] = time
    
    
```

```python
df_r['Марка'] = df_r['Марка'].str[len('Марка стали'):].astype('int')
```

## Удаление ненужных параметров

```python
df_t = df_t.drop(['дата_завалки', 'дата_вывалки'], axis=1)
df_r = df_r.drop(['Время_обработки'], axis=1)
```

# Создание единого датасета

```python
df_v = pd.get_dummies(df_v,  columns=['материал_валка'])
```

```python
df_r['Толщина_х_Ширина'] = df_r['Толщина'] * df_r['Ширина']
```

```python
df_rl = pd.DataFrame()
```

```python
df_rl['sum_tol'] = df_r.groupby(['rul_id'])['Толщина'].sum()
df_rl['Толщина_х_Ширина'] = df_r.groupby(['rul_id'])['Толщина'].median()
df_rl['median_time'] = df_r.groupby(['rul_id'])['time'].median()
df_rl['sum_mass'] = df_r.groupby(['rul_id'])['Масса'].sum()
df_rl['counts'] = df_r.groupby(['rul_id'])['Масса'].count()
```

```python
counter_df_train=df_r.groupby(['rul_id','Марка'])['Толщина_х_Ширина'].sum()
cat_counts_train=counter_df_train.reset_index().pivot(index='rul_id', \
                                                      columns='Марка',values='Толщина_х_Ширина')
mass = df_r.groupby(['rul_id',])['Толщина_х_Ширина'].sum()
cat_counts_train=cat_counts_train.fillna(0)
cat_counts_train.columns=['Marka_tol_sh_sum_'+str(i) for i in cat_counts_train.columns]
for i in cat_counts_train.columns:
    cat_counts_train[i] /= mass
df_rl = pd.merge(df_rl, cat_counts_train, on='rul_id')
```

```python
counter_df_train=df_r.groupby(['rul_id','Марка'])['time'].sum()
cat_counts_train=counter_df_train.reset_index().pivot(index='rul_id', \
                                                      columns='Марка',values='time')
mass = df_r.groupby(['rul_id',])['time'].sum()
cat_counts_train=cat_counts_train.fillna(0)
cat_counts_train.columns=['Marka_time_sum_'+str(i) for i in cat_counts_train.columns]
for i in cat_counts_train.columns:
    cat_counts_train[i] /= mass
df_rl = pd.merge(df_rl, cat_counts_train, on='rul_id')
```

```python
counter_df_train=df_r.groupby(['rul_id','Марка'])['Масса'].sum()
cat_counts_train=counter_df_train.reset_index().pivot(index='rul_id', \
                                                      columns='Марка',values='Масса')
mass = df_r.groupby(['rul_id',])['Масса'].sum()
cat_counts_train=cat_counts_train.fillna(0)
cat_counts_train.columns=['Marka_sum_'+str(i) for i in cat_counts_train.columns]
for i in cat_counts_train.columns:
    cat_counts_train[i] /= mass
df_rl = pd.merge(df_rl, cat_counts_train, on='rul_id')
```

```python
counter_df_train=df_r.groupby(['rul_id','Марка'])['Толщина'].median()
cat_counts_train=counter_df_train.reset_index().pivot(index='rul_id', \
                                                      columns='Марка',values='Толщина')
cat_counts_train=cat_counts_train.fillna(0)
cat_counts_train.columns=['Marka_tol_median_'+str(i) for i in cat_counts_train.columns]
df_rl = pd.merge(df_rl, cat_counts_train, on='rul_id')
```

```python
dct = {}
iz = df_t['износ'].values
v =  df_t['номер_валка'].values
valu = []
for i in range(len(v)):
    if v[i] not in dct.keys():
        valu.append(0)
        dct[v[i]] = 0
    else:
        valu.append(dct[v[i]])
        dct[v[i]] += iz[i]
df_t['доп_износ'] = valu
```

```python
df_t = pd.merge(df_t, df_rl, on='rul_id')
```

```python
df_t = pd.merge(df_t, df_v, on='номер_валка')
```

```python
df_t['time/mass'] = df_t['время']/df_t['sum_mass']
```

# Обучение

```python
X = df_t.drop(['rul_id', 'износ', 'номер_валка'], axis=1)
y = df_t['износ']
```

## Оценка

```python
from sklearn.model_selection import train_test_split
import catboost
```

```python
X1, xc, y1, yc = train_test_split(X, y, test_size=0.8, random_state=42)
X2, xc, y2, yc = train_test_split(xc, yc, test_size=0.75, random_state=42)
X3, xc, y3, yc = train_test_split(xc, yc, test_size=0.66, random_state=42)
X4, X5, y4, y5 = train_test_split(xc, yc, test_size=0.5, random_state=42)
```

```python
scores = []
```

```python
model = catboost.CatBoostRegressor(iterations=5000, depth=6, l2_leaf_reg=2, cat_features=[0, 1])
X_tr = X2.append(X3).append(X4).append(X5)
y_tr = y2.append(y3).append(y4).append(y5)
model.fit(X_tr, y_tr)
y_pred1 = model.predict(X1)
scores.append([(sum(abs(y1 - y_pred1))/len(y_pred1)), (sum(abs(y1 - y_pred1))/len(y_pred1))/(sum(y1)/len(y_pred1))])
##
model = catboost.CatBoostRegressor(iterations=5000, depth=6, l2_leaf_reg=2, cat_features=[0, 1])
X_tr = X1.append(X3).append(X4).append(X5)
y_tr = y1.append(y3).append(y4).append(y5)
model.fit(X_tr, y_tr)
y_pred1 = model.predict(X2)
scores.append([(sum(abs(y2 - y_pred1))/len(y_pred1)), (sum(abs(y2 - y_pred1))/len(y_pred1))/(sum(y2)/len(y_pred1))])
##
model = catboost.CatBoostRegressor(iterations=5000, depth=6, l2_leaf_reg=2, cat_features=[0, 1])
X_tr = X2.append(X1).append(X4).append(X5)
y_tr = y2.append(y1).append(y4).append(y5)
model.fit(X_tr, y_tr)
y_pred1 = model.predict(X3)
scores.append([(sum(abs(y3 - y_pred1))/len(y_pred1)), (sum(abs(y3 - y_pred1))/len(y_pred1))/(sum(y3)/len(y_pred1))])
##
model = catboost.CatBoostRegressor(iterations=5000, depth=6, l2_leaf_reg=2, cat_features=[0, 1])
X_tr = X2.append(X3).append(X1).append(X5)
y_tr = y2.append(y3).append(y1).append(y5)
model.fit(X_tr, y_tr)
y_pred1 = model.predict(X4)
scores.append([(sum(abs(y4 - y_pred1))/len(y_pred1)), (sum(abs(y4 - y_pred1))/len(y_pred1))/(sum(y4)/len(y_pred1))])
##
model = catboost.CatBoostRegressor(iterations=5000, depth=6, l2_leaf_reg=2, cat_features=[0, 1])
X_tr = X2.append(X3).append(X4).append(X1)
y_tr = y2.append(y3).append(y4).append(y1)
model.fit(X_tr, y_tr)
y_pred1 = model.predict(X5)
scores.append([(sum(abs(y5 - y_pred1))/len(y_pred1)), (sum(abs(y5 - y_pred1))/len(y_pred1))/(sum(y5)/len(y_pred1))])
```
